
### Answer:
If any exception or error occurs during runtime of a program it will crash. To prevent this "try except" block is used. Inside try block we write our code which is supposed to run and codes inside except block will run in case any error occurs inside try block.    

```python
# Answer
try:
    # Your code here 
    print(10/0)
    
except Exception as e:
    # Print the error to console
    print(f"Error occured - {e}")
```


### Answer:
If any matching error is not found in except block the error will not handled and the program will crash.

```python
try:
    print(10/0)
except IndexError as e:
    print("Index error")
    
```


### Answer:
Using a bare exception block will handle every exception in same way. Using exception type will hadnle different exceptions in different ways.

```python
# Answer:
x = 10
y = '0'
try:
    print(x/y)
    
except TypeError as e:
    print(e)
    
    try:
        print(int(x)/int(y))
        
    except ZeroDivisionError as e:
        print(e)
```

```python
def multiple(arr):
    try:
        for i in range(10):
            _ = arr[i]/5
        
    except TypeError as e:
        print(e)

    except IndexError as e:
        print(e)

multiple([10, 20, 30, '14', 15])
multiple([10, 20, 30, 14, 15])
```

```python
def div(a,b):
    try:
        return a/b
    except ZeroDivisionError as e:
        print("Division by 0 not allowed")
div(10,0)
```

```python
def convert(string):
    try:
        return int(string)
    except ValueError as e:
        print("Only numerical inputs accepted")
convert("asdasd")
```
