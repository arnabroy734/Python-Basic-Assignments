## Question - 3


#### Logical operators are used to check conditional statements. Three logical operators are as follows:
##### 1. and  - it is used to check whether all the conditions are true
##### 2. or - it is used to check whether either of the condition is true
##### 3. not  - it is used to check whether inversion of the condition is true

```python
print( (10>5) and (5>10) )
```

```python
print( (10>5) or (5>10) )
```

```python
print( not (5>10) )
```

## Question - 4


#### x<<y operators shifts the bits of x to left y times and fills zeroes. Each shift is equivalent to mulplication by 2

```python
def dec_2_bin(number):
    print(f"Decimal - {number}")
    print(f"Binary - {bin(number).replace('0b','')}")
```

```python
dec_2_bin(16)
```

```python
dec_2_bin(16<<2)
```

#### x>>y operators shifts the bits of x to right y times. Each shift is equivalent to division by 2

```python
dec_2_bin(100)
```

```python
dec_2_bin(100>>3)
```
